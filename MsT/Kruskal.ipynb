•Optimal Substructure:the optimal solution to aproblem incorporates the  optimal solution to subproblem(s) 
•Greedy choice property:locally optimal choices lead to aglobally optimal solution. We can see how these properties can be applied to the MST problem

### 1. Log Return

According to the Seminal Paper by Mantegna and Stanbley, the log returns of a time-series are modelled by a Levy distribution.
A mixture of Gaussian Distributions with $\mathcal{X}^{2}$- distributions in terms of variance gives a Levy distribution.
The logarithmic return are the Logarithmic Likelihood ratio under a gaussian null model.
A return at a time $(t)$ is represented by the formula $\Delta(t) =\log(t2)-\log(t1) $


log return fluctuations are constantly affected by avalanches of market orders which have to do with speculative trends, and are time localized events.

It is important to note that asset returns models have to take into account the concept of subordination, we will assume that the logreturns take that into account and is trivial for discussion in this notebook.

As price denoted by $ P_{i} $ is known as a tochastic variable it follows that an inflation has a direct impact on its actual fair market value, with inflation rate $ I_{r} > 0 $  , a need arise for an investor to modify log returns to real-log returns. 
We are constrained to using nominal log returns as opposed to the modified returns due to the fact that our tickers choice is not restricted to a specific country. We can remove this bottleneck by making use of the global inflation average however this is prone to accuracy errors. 

### 1.1 Log Return and Morden Portfolio theory

Modern Portfolio Theory assumes that investors are risk averse, where the risk is measured by the variance of asset price. It basically describes the ”trade off” between return and risk.
Investors who want higher return must accept higher risk, but different people may have their own risk toleration, which lead to different investment strategies, forming the so-called Efficient Frontier. Under the model, we can represent the expected portfolio return as

${\mathbb{E}}(R_{p})= \sum\limits_{i=1}^n W_{i} {\mathbb{E}}(R_{i})$

from the the above $ W_{i} $ is the weight of the asset and $ R_{i} $ is the corresponding asset return

Given the return of an individual asset and its corresponding variance you need to consider your risk appetite in forming a portfolio.

# 2.Diversification

For diversification purposes in selecting a portfolio by principle an investor would select a portfolio of assets that moves into different directions.

## 2.1 Hedge Considerations

A hedge is an investment to reduce the risk of adverse price movements in an asset. Normally, a hedge consists of taking an offsetting position in a related security. This relationship can be studied by using the adjacent matrix if the values offset to zero implies strong positive correlation, as the values range above zero towards a higher number in this note book we will be using a scaling limit of 2 based on the distance observations.

**Example: KRTX Union NIO**

Given the two tickers KRTX and NIO, from the clustermap we can observe that they are negatively correlated, in comparison to their distances (use the code  **ks.distance()['KRTX']** to find the values) the value is **1.820687**, this represent how uncorrelated they are representing a good portfolio given hedging consideretions.
This will results to a approximately  perfect hedge defined as one that eliminates all risk in a position or portfolio. In other words, the hedge is 100% inversely correlated to the vulnerable asset.

```python
#Visualizing distances
ks.distance()
```

```python
# Plotting the heatmap
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(ks.distance(), ax = ax, cbar_kws={'label': 'Distances Distritutions'})

ax.set_title('Heatmap of distance distributions')
plt.show()
```
